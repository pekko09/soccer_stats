```python
from moduls.fixtures import MatchFixtures

bvb = MatchFixtures(teamid = 7, year_season = 2025)

for c in {'bl1', 'ucl', 'dfb'}:
    bvb.get_matches(comp=c)

bvb.matches.sort(key=lambda x: x['matchDateTime'])
```

```python
comp_labels = {'bl1': 'BL', 'ucl': 'CL', 'dfb': 'DFB'}

for match in bvb.matches:
    comp = match['leagueShortcut']
    home = match['team1']['teamName']
    away = match['team2']['teamName']
    date = match['matchDateTime']

    print(f"{date} {comp_labels[comp]} {home} - {away}")
```
